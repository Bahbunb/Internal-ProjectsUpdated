# Webscraping from wikipedia

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Demographics_of_Singapore#Household_income'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
# Find all tables
tables = soup.find_all('table')

# Check each table for the caption text
for i, table in enumerate(tables):
    caption = table.find('caption')
    if caption and 'Household income from work' in caption.get_text():
        print(f"Table found at index {i}:")
        print(caption.get_text())
        # Optionally, print the table HTML or specific rows
        print(table.prettify())
        break
else:
    print("Table not found!")
```

```python
table = soup.find_all('table')[30]
```

```python
print(table)
```

```python
income_data = table.find_all('th')
```

```python
income_data
```

```python
income_table_data = [title.text.strip() for title in income_data]

print(income_table_data)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = income_table_data)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df
```
