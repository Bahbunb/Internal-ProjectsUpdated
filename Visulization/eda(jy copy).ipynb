```python
# You will now be able to import required libraries
import pandas as pd         # data manipulation & analysis
import numpy as np          # numerical & mathematical operations
import matplotlib.pyplot as plt         # for creating visualisations
import seaborn as  sns          # for more advanced visualisations 
```

```python
df = pd.read_csv('updated_dataset.csv') 
```

```python
df.head()
```

```python
#to view the unique elements in the months, town, flat_types column
print(df['month'].unique())
print(df['town'].unique())
print(df['flat_type'].unique())
```

```python
#to count how many units in each town in the dataset ... ?

town_counts = df['town'].value_counts()
print(town_counts)
```

```python
table = town_counts.reset_index()
table.columns = ['Town', 'Count']  # Rename columns for clarity

# Display the table
print(table)
```

```python
# Plot the data
plt.figure(figsize=(14, 6))
town_counts.sort_values(ascending=False).plot(kind='bar', color='blue')
plt.title('Resale Flats by Town', fontsize=16)
plt.xlabel('Town', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate town names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```

```python
# Calculate the average resale price by town and flat type
average_prices = df.groupby(['town', 'flat_type'])['resale_price'].mean().reset_index()

# Rename columns for clarity
average_prices.columns = ['Town', 'Flat Type', 'Average Resale Price']

# Sort the table for better readability
average_prices = average_prices.sort_values(by=['Town', 'Flat Type']).reset_index(drop=True)

# Display the first few rows of the table
pd.set_option('display.max_rows', None)
print(average_prices)
```

```python
average_prices['Average Resale Price'] = average_prices['Average Resale Price'].round(2)

# Display the updated table
print(average_prices)
```

```python
# Add a new column for price per square foot (1 sqm = 10.7639 sq ft)
df['price_per_sqm'] = df['resale_price'] / (df['floor_area_sqm'] * 10.7639)

# Calculate the average price per square foot for each town
average_price_per_sqm = df.groupby('town')['price_per_sqm'].mean().reset_index()

# Rename columns for clarity
average_price_per_sqm.columns = ['Town', 'Average Price Per Sqm']

# Round to 2 decimal places for readability
average_price_per_sqm['Average Price Per Sqm'] = average_price_per_sqm['Average Price Per Sqm'].round(2)

# Sort the table by town for better readability
# Sort by 'Average Price Per Sqft' in descending order
average_price_per_sqm = average_price_per_sqm.sort_values(by='Average Price Per Sqm', ascending=False).reset_index(drop=True)


# Display the resulting table
plt.figure(figsize=(14, 6))
plt.bar(average_price_per_sqm['Town'], average_price_per_sqm['Average Price Per Sqm'], color='skyblue')

# Add labels and title
plt.title('Average Price Per Square Meter by Town', fontsize=16)
plt.xlabel('Town', fontsize=14)
plt.ylabel('Average Price Per Sqm ($)', fontsize=14)
plt.xticks(rotation=45, ha='right')  # Rotate town names for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust layout and show the chart
plt.tight_layout()
plt.show()
```
